# Najprostszy model

* Wykorzystanie tylko atrybutów numerycznych
* Porównanie regresji liniowej z modelem losowym
  * U(0,5) i U(4,5)

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
import numpy as np
%matplotlib inline

DATA_VERSION = "v2"

calendar = pd.read_csv(f'../data/{DATA_VERSION}/raw/calendar.csv')
listings = pd.read_csv(f'../data/{DATA_VERSION}/raw/listings.csv')
reviews = pd.read_csv(f'../data/{DATA_VERSION}/raw/reviews.csv')
```

```python
listings.info()
```

```python
rating_columns = [
    "review_scores_rating",
    "review_scores_accuracy",
    "review_scores_cleanliness",
    "review_scores_checkin",
    "review_scores_communication",
    "review_scores_location",
    "review_scores_value",
]
listings["avg_rating"] = listings[rating_columns].mean(axis=1)
```

```python
listings["price"] = listings["price"].str.replace("$", "").str.replace(",", "").astype(float)
```

```python
listings["avg_rating"].describe()
```

```python
drop_columns = [
    "id",
    "listing_url",
    "scrape_id",
    "last_scraped",
    "source",
    "name",
    "description",
    "neighborhood_overview",
    "picture_url",
    "host_id",
    "host_url",
    "host_name",
    "host_since",
    "host_location",
    "host_about",
    "host_response_time",
    "host_response_rate",
    "host_acceptance_rate",
    "host_is_superhost",
    "host_thumbnail_url",
    "host_picture_url",
    "host_neighbourhood",
    "host_verifications",
    "host_has_profile_pic",
    "host_identity_verified",
    "neighbourhood",
    "neighbourhood_cleansed",
    "neighbourhood_group_cleansed",
    "latitude",
    "longitude",
    "property_type",
    "room_type",
    "bathrooms_text",
    "amenities",
    "minimum_nights",
    "maximum_nights",
    "calendar_updated",
    "has_availability",
    "calendar_last_scraped",
    "first_review",
    "last_review",
    "review_scores_rating",
    "review_scores_accuracy",
    "review_scores_cleanliness",
    "review_scores_checkin",
    "review_scores_communication",
    "review_scores_location",
    "review_scores_value",
    "license",
    "instant_bookable",
    "reviews_per_month",
]
listings = listings.drop(columns=drop_columns)
```

```python
listings.head()
```

```python
print(len(listings))
print(listings["avg_rating"].isna().sum())

listings.dropna(inplace=True)
print(len(listings))
```

```python
listings
```

```python
def linear_regression_results(listings):
    print(f"Rows: {len(listings)}")
    Y = listings["avg_rating"]
    X = listings.drop(columns=["avg_rating"])

    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
    model = LinearRegression()
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    print("Linear regression")
    print(f"Mean Squared Error: {mse:.2f}")
    print(f"R-squared: {r2:.2f}")

    print("-"*30)

    print("Random model - U(0,5)")
    y_random_pred = np.random.rand(len(y_test)) * 5
    mse_random = mean_squared_error(y_test, y_random_pred)
    r2_random = r2_score(y_test, y_random_pred)
    print(f"Mean Squared Error: {mse_random:.2f}")
    print(f"R-squared: {r2_random:.2f}")

    print("-"*30)

    print("Random model - U(4,5)")
    y_random_pred = np.random.rand(len(y_test)) + 4
    mse_random = mean_squared_error(y_test, y_random_pred)
    r2_random = r2_score(y_test, y_random_pred)
    print(f"Mean Squared Error: {mse_random:.2f}")
    print(f"R-squared: {r2_random:.2f}")
```

```python
# Use raw data with only numeric columns
linear_regression_results(listings)
```
